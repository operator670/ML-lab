```python
from sklearn import preprocessing
import numpy as np
```

```python
lb = preprocessing.LabelEncoder()
```

```python
x1 = lb.fit_transform(df['no'])
x2 = lb.fit_transform(df['pc'])
x3 = lb.fit_transform(df['wbc'])
x4 = lb.fit_transform(df['mc'])
x5 = lb.fit_transform(df['ast'])
x6 = lb.fit_transform(df['bc'])
x7 = lb.fit_transform(df['ldh'])
yxc = lb.fit_transform(df['diagnosis'])
```

```python
features = zip(x2,x3,x4,x5,x6,x7)
features = np.array(list(features))
```

```python
features
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,yxc,test_size=0.2)
```

```python
mnb = MultinomialNB()
```

```python
y_pred = mnb.fit(x_train,y_train).predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```
